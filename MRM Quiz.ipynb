```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'airbnb_data.csv' with the actual filename)
df = pd.read_csv("listings_detailed.csv")

# Count room type occurrences
room_type_counts = df["room_type"].value_counts()

# Count property type occurrences
property_type_counts = df["property_type"].value_counts().head(10)  # Show top 10 property types

# Plot room type distribution
plt.figure(figsize=(8, 5))
room_type_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribution of Room Types in Airbnb Listings")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Plot property type distribution
plt.figure(figsize=(10, 5))
property_type_counts.plot(kind="bar", color="lightcoral", edgecolor="black")
plt.title("Top 10 Property Types in Airbnb Listings")
plt.xlabel("Property Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Display raw counts
print("Room Type Distribution:")
print(room_type_counts)
print("\nTop 10 Property Types:")
print(property_type_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'airbnb_data.csv' with the actual filename)
df = pd.read_csv("listings_detailed.csv")

# Count listings per host
host_listings_count = df["host_id"].value_counts()

# Summary statistics
single_listing_hosts = (host_listings_count == 1).sum()
multi_listing_hosts = (host_listings_count > 1).sum()
max_listings_by_a_host = host_listings_count.max()

# Plot distribution of listings per host (log scale for better visualization)
plt.figure(figsize=(10, 5))
sns.histplot(host_listings_count, bins=50, kde=True, log_scale=(True, False), color="teal")
plt.title("Distribution of Listings per Host")
plt.xlabel("Number of Listings per Host (Log Scale)")
plt.ylabel("Number of Hosts")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Show summary insights
print(f"Total Hosts: {len(host_listings_count)}")
print(f"Hosts with a single listing: {single_listing_hosts} ({(single_listing_hosts / len(host_listings_count)) * 100:.2f}%)")
print(f"Hosts with multiple listings: {multi_listing_hosts} ({(multi_listing_hosts / len(host_listings_count)) * 100:.2f}%)")
print(f"Maximum listings owned by a single host: {max_listings_by_a_host}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from langdetect import detect, DetectorFactory
import scipy.stats as stats
import statsmodels.formula.api as smf

# Ensure language detection is consistent
DetectorFactory.seed = 0

# Load dataset (replace 'airbnb_data.csv' with actual file)
df = pd.read_csv("listings_detailed.csv")

# Function to detect language
def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"

# Apply language detection on the 'description' column
df["language"] = df["description"].astype(str).apply(detect_language)

# Filter for English and Spanish listings only
df_lang_filtered = df[df["language"].isin(["en", "es"])]

# Convert price column to numeric
df_lang_filtered["price"] = df_lang_filtered["price"].replace('[\$,]', '', regex=True).astype(float)

# Compare average price between English & Spanish listings
avg_price_by_lang = df_lang_filtered.groupby("language")["price"].mean()

# Box plot for price distribution by language
plt.figure(figsize=(8, 5))
sns.boxplot(x="language", y="price", data=df_lang_filtered, palette={"en": "lightblue", "es": "lightcoral"})
plt.title("Price Distribution by Listing Language (Tourist vs Local)")
plt.xlabel("Language (en=English, es=Spanish)")
plt.ylabel("Price ($)")
plt.ylim(0, df_lang_filtered["price"].quantile(0.95))  # Limit outliers
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Analyze price differences by neighborhood
price_by_lang_neighborhood = df_lang_filtered.groupby(["neighbourhood_cleansed", "language"])["price"].mean().unstack()

# Plot price difference in top neighborhoods
top_neighborhoods = price_by_lang_neighborhood.dropna().sort_values(by="en", ascending=False).head(10)
top_neighborhoods.plot(kind="bar", figsize=(12, 6), color=["lightblue", "lightcoral"], edgecolor="black")
plt.title("Average Price by Listing Language in Top Neighborhoods")
plt.xlabel("Neighborhood")
plt.ylabel("Average Price ($)")
plt.legend(["English Listings (Tourist)", "Spanish Listings (Local)"])
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Perform t-test to compare prices between English and Spanish listings
english_prices = df_lang_filtered[df_lang_filtered["language"] == "en"]["price"]
spanish_prices = df_lang_filtered[df_lang_filtered["language"] == "es"]["price"]

t_stat, p_value = stats.ttest_ind(english_prices, spanish_prices, equal_var=False)  # Welchs t-test

# Print t-test results
print(f"T-test results: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

# Regression model: Does language predict price?
df_lang_filtered = df_lang_filtered.dropna(subset=["price", "language"])  # Drop missing values

# Encode language as a binary variable (English = 1, Spanish = 0)
df_lang_filtered["language_encoded"] = df_lang_filtered["language"].apply(lambda x: 1 if x == "en" else 0)

# Fit a linear regression model
model = smf.ols("price ~ language_encoded + C(neighbourhood_cleansed)", data=df_lang_filtered).fit()

# Display regression summary
print(model.summary())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

# Load dataset
df = pd.read_csv("listings_detailed.csv")

# Convert price column to numeric (remove "$" and convert to float)
df["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)

# Convert 'host_identity_verified' to numeric (1 = Verified, 0 = Not Verified)
df["host_identity_verified"] = df["host_identity_verified"].map({"t": 1, "f": 0})

# Ensure bathrooms column is numeric
df["bathrooms"] = pd.to_numeric(df["bathrooms"], errors='coerce')

# Select relevant columns
selected_columns = ["price", "accommodates", "host_identity_verified", "bathrooms", "bedrooms", "beds", "review_scores_rating"]
df_selected = df[selected_columns]

# Check for missing values
print("Missing Values:\n", df_selected.isnull().sum())

# Fill missing numeric values with median
for col in ["bathrooms", "bedrooms", "beds", "review_scores_rating"]:
    df_selected[col].fillna(df_selected[col].median(), inplace=True)

# Fill missing 'host_identity_verified' with mode
df_selected["host_identity_verified"].fillna(df_selected["host_identity_verified"].mode()[0], inplace=True)

# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df_selected.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: Factors Affecting Price")
plt.show()

# Multiple Linear Regression Model
formula = "price ~ accommodates + host_identity_verified + bathrooms + bedrooms + beds + review_scores_rating"
model = smf.ols(formula, data=df_selected).fit()

# Display regression summary
print(model.summary())
```
