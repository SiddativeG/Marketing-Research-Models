```python
# Step 1: Install required libraries
!pip install textblob wordcloud
!pip install nltk
!pip install scikit-learn

# Step 2: Import necessary libraries
import pandas as pd
import re
import matplotlib.pyplot as plt
from textblob import TextBlob
from google.colab import files

# For content analysis
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Download NLTK stopwords if needed
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Step 3: Upload your Excel file (use the file chooser w

# Step 4: Read the Excel file into a DataFrame
df = pd.read_csv('Reviews 500.csv')

# Display the first few rows to verify the data
print("Data Preview:")
print(df.head())

####################################
# Sentiment Analysis using TextBlob
####################################

# Define a function to compute sentiment polarity
def analyze_sentiment(review):
    blob = TextBlob(str(review))
    return blob.sentiment.polarity

# Apply sentiment analysis on the 'Text' column
df['Sentiment'] = df['Text'].apply(analyze_sentiment)

# Optionally, categorize sentiment
def sentiment_category(polarity):
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment_Category'] = df['Sentiment'].apply(sentiment_category)

print("\nData with Sentiment Analysis:")
print(df[['Text', 'Sentiment', 'Sentiment_Category']].head())

####################################
# Content Analysis
####################################

# Step 5: Text Cleaning Function
def clean_text(text):
    text = str(text).lower()                           # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)                # Remove punctuation and non-alphabetic characters
    text = re.sub(r'\s+', ' ', text).strip()            # Remove extra whitespace
    return text

# Apply cleaning to the 'Text' column
df['Cleaned_Text'] = df['Text'].apply(clean_text)

# Step 6: Word Frequency Analysis using CountVectorizer
# You can supply additional stopwords if desired
stop_words = stopwords.words('english')
vectorizer = CountVectorizer(stop_words=stop_words)
X = vectorizer.fit_transform(df['Cleaned_Text'])

# Sum the occurrences of each word
word_counts = X.toarray().sum(axis=0)
words = vectorizer.get_feature_names_out()
word_freq = dict(zip(words, word_counts))

# Display top 10 most frequent words
top_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nTop 10 Most Frequent Words:")
for word, count in top_words:
    print(f"{word}: {count}")

# Step 7: Word Cloud Visualization
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Reviews")
plt.show()

# Step 8: (Optional) Topic Modeling using Latent Dirichlet Allocation (LDA)
# Setting the number of topics; you can adjust n_components based on your data
n_topics = 3
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(X)

# Function to display the top words for each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"\nTopic {topic_idx + 1}:")
        top_features_ind = topic.argsort()[:-no_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        print(" ".join(top_features))

print("\nTopics Identified by LDA:")
display_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
# Step 1: Install required libraries
!pip install prophet openpyxl  # openpyxl is used for reading Excel files

# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from google.colab import files

# Step 4: Read the Excel file into a DataFrame
df = pd.read_csv('Superstore - store analysis.csv', encoding='latin-1')

# Convert the 'Order Date' column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', errors='coerce')

######################################
# Daily Forecast with Prophet
######################################

# Aggregate daily sales
daily_sales = (
    df.groupby('Order Date')['Sales']
      .sum()
      .reset_index()
      .rename(columns={'Order Date': 'ds', 'Sales': 'y'})
)

# Create a sample holiday dataframe (customize as needed)
holidays = pd.DataFrame({
    'holiday': 'us_holiday',
    'ds': pd.to_datetime([
        '2014-01-01', '2014-07-04', '2014-12-25',
        '2015-01-01', '2015-07-04', '2015-12-25',
        '2016-01-01', '2016-07-04', '2016-12-25'
    ]),
    'lower_window': 0,
    'upper_window': 1,
})

# Initialize and fit the Prophet model with holiday effects
m_daily = Prophet(
    holidays=holidays,
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False  # Typically, daily seasonality may be too noisy for sales
)
m_daily.fit(daily_sales)

# Create a future dataframe for the next 365 days (daily forecast)
future_daily = m_daily.make_future_dataframe(periods=365)
forecast_daily = m_daily.predict(future_daily)

# Plot the daily forecast and its components
fig1 = m_daily.plot(forecast_daily)
plt.title("Daily Sales Forecast")
plt.show()

fig2 = m_daily.plot_components(forecast_daily)
plt.show()

######################################
# Weekly Forecast with Prophet
######################################

# Aggregate sales data by week
weekly_sales = df.copy()
# Convert 'Order Date' to the week period (using the start of the week)
weekly_sales['Order Week'] = df['Order Date'].dt.to_period('W').apply(lambda r: r.start_time)
weekly_sales = (
    weekly_sales.groupby('Order Week')['Sales']
                .sum()
                .reset_index()
                .rename(columns={'Order Week': 'ds', 'Sales': 'y'})
)

# Initialize and fit the Prophet model for weekly data (turn off weekly seasonality as data is weekly)
m_weekly = Prophet(
    yearly_seasonality=True,  # You may still see yearly trends
    weekly_seasonality=False
)
m_weekly.fit(weekly_sales)

# Create future dataframe for the next 52 weeks
future_weekly = m_weekly.make_future_dataframe(periods=52, freq='W')
forecast_weekly = m_weekly.predict(future_weekly)

# Plot the weekly forecast and its components
fig3 = m_weekly.plot(forecast_weekly)
plt.title("Weekly Sales Forecast")
plt.show()

fig4 = m_weekly.plot_components(forecast_weekly)
plt.show()

######################################
# Monthly Forecast with Prophet
######################################

# Aggregate sales data by month
monthly_sales = df.copy()
# Convert 'Order Date' to the month period (using the start of the month)
monthly_sales['Order Month'] = df['Order Date'].dt.to_period('M').apply(lambda r: r.start_time)
monthly_sales = (
    monthly_sales.groupby('Order Month')['Sales']
                 .sum()
                 .reset_index()
                 .rename(columns={'Order Month': 'ds', 'Sales': 'y'})
)

# Initialize and fit the Prophet model for monthly data (turn off weekly seasonality)
m_monthly = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False
)
m_monthly.fit(monthly_sales)

# Create future dataframe for the next 12 months
future_monthly = m_monthly.make_future_dataframe(periods=12, freq='M')
forecast_monthly = m_monthly.predict(future_monthly)

# Plot the monthly forecast and its components
fig5 = m_monthly.plot(forecast_monthly)
plt.title("Monthly Sales Forecast")
plt.show()

fig6 = m_monthly.plot_components(forecast_monthly)
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from google.colab import files

# Upload file
df = pd.read_csv('MMM brand data.csv')

# Define Adstock Function
def adstock(series, decay=0.5):
    result = np.zeros(len(series))
    result[0] = series.iloc[0]
    for t in range(1, len(series)):
        result[t] = series.iloc[t] + decay * result[t-1]
    return result

# Apply Adstock transformation
media_channels = ["TV_Spend", "Radio_Spend", "Print_Spend", "Search_Spend", "YouTube_Spend", "Social Media_Spend"]
for channel in media_channels:
    df[channel + "_Adstock"] = adstock(df[channel])

# Apply Saturation Effect using a Hill Function ( = 0.8, S = max spend)
beta = 0.8
S = df[media_channels].max().max()
for channel in media_channels:
    df[channel + "_Saturation"] = (df[channel] ** beta) / (df[channel] ** beta + S ** beta)

# Log transform Price
df["Log_Price"] = np.log(df["Price"])

# Interaction of Promotion & Price
df["Promotion_Price_Interaction"] = df["Promotion"] * df["Log_Price"]

# Define independent variables
X = df[["Log_Price", "Promotion", "Promotion_Price_Interaction"]]
for channel in media_channels:
    X[channel + "_Adstock"] = df[channel + "_Adstock"]
    X[channel + "_Saturation"] = df[channel + "_Saturation"]

# Define dependent variable
y = df["Sales"]

# Add constant for regression
X = sm.add_constant(X)

# Run OLS Regression Model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())
```
